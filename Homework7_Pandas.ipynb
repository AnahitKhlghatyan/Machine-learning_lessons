```python
# 8. Կարդալ imdb_top_1000.csv ֆայլը։ Այստեղ կտեսնենք IMDB կայքի տվյալների բազայի լավագույն 1000 ֆիլմերի ցանկը։
```

```python
df_f = pd.read_csv(r'./imdb_top_1000.csv')
df_f
```

```python
# 1. Ստուգել դատասեթի առաջին 8 տողը
```

```python
df_f.head(8)
```

```python
# 2. Գտնել բազայում ներառված ամենահին ֆիլմը և տպել դրա ռեժիսորին և ժանրը
```

```python
df_f['Released_Year'] = df_f['Released_Year'].str.replace('PG','1995')
df_f['Released_Year'] = df_f['Released_Year'].astype(int)
df_f['Released_Year'].dtype
```

```python
group = df_f.groupby('Released_Year')
group['Genre','Director'].min()
```

```python
# 3. Գտնել ամենամեծ քանակությամբ ֆիլմեր ունեցող ռեժիսորին (կարող եք խմբավորել ըստ ռեժիսորների)
```

```python
group = df_f.groupby('Director')
dirc = group.size().sort_values().tail(1)
dirc
```

```python
#4. Հաշվել այդ ռեժիսորի ֆիլմերի միջին, առավելագույն և նվազագույն ռեյտինգները։
```

```python
alf = group.get_group('Alfred Hitchcock')
ag = alf.agg({'IMDB_Rating': ['min', 'mean', 'max']})
ag
```

```python
# 5․ Գտնել այն ռեժիսորներին, ովքեր դատասեթում ունեն 1-ից ավել ֆիլմ և գտնել նվազագույն միջին ռեյտինգ ունեցողին։
```

```python
grp = df_f.groupby('Director').agg(Count = ('Director', 'count'), Rating = ('IMDB_Rating', 'mean'))
filt = grp['Count'] > 2
grp[filt]
```

```python
grp[filt].sort_values(by='Rating')
```

```python
# 6. Տպել Meta score սյան միջին և միջին քառակուսային շեղման արժեքները բոլոր ռեժիսորների համար
```

```python
ag = group.agg({'Meta_score': ['mean','std']})
ag
```

```python
# 7. Դատասեթում բացակայող արժեքները լրացնել այդ սյան միջին արժեքով։ Եթե սյունը թվային տիպ չունի, պարզապես ջնջել այդտեղ բացակայող արժեքներին համապատասխան տողերը
```

```python
df_f['Gross'] = df_f['Gross'].str.replace(',','')
df_f['Gross'] = df_f['Gross'].astype(float)
```

```python
df_f.fillna(df_f.mean(),inplace=True)
```

```python
df_f.dropna(inplace=True)
df_f
```

```python
# 8. Սորտավորել դատասեթը ըստ տարեթվերի և այն պահել նոր ֆայլի մեջ
```
